```python
import pandas as pd
import requests
import json
import os

import matplotlib.pyplot as plt
import seaborn as sns
```

# Import and Clean up AAPL data set
## Convert the 'Vol.' column to represent a numeric value


```python
# Import AAPL Historical Data from CSV
AAPL_Historical_Data_W = pd.read_csv(f"../Datasets/AAPL_Historical_Data_W.csv")
AAPL_Historical_Data_W

# Convert Date column to datetime format
AAPL_Historical_Data_W['Date'] = pd.to_datetime(AAPL_Historical_Data_W['Date'], format='%m/%d/%Y')
AAPL_Historical_Data_W

# Show Data Types
AAPL_Historical_Data_W.dtypes

# In 'Vol.' column there are 'M' and 'B' values, we need to convert them to numbers
# Remove M and multiply by 1,000,000
# Remove B and multiply by 1,000,000,000

# 1. Handle missing values in the "Vol." column
AAPL_Historical_Data_W['Vol.'] = AAPL_Historical_Data_W['Vol.'].fillna('0')

# 2. Convert "Vol." column to numeric format
AAPL_Historical_Data_W['Vol.'] = AAPL_Historical_Data_W['Vol.'].str.replace('B', 'e9').str.replace('M', 'e6').map(pd.eval).astype(int)

AAPL_Historical_Data_W

# Show days with the highest trading volume
AAPL_Historical_Data_W.nlargest(10, 'Vol.')



```

# Plot AAPL Price and Volume since IPO

```python
# Convert 'Date' column to datetime format (if not already)
AAPL_Historical_Data_W['Date'] = pd.to_datetime(AAPL_Historical_Data_W['Date'])

# Plotting Price over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Price', data=AAPL_Historical_Data_W)
plt.title('AAPL Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()

# Plotting Volume over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Vol.', data=AAPL_Historical_Data_W)
plt.title('AAPL Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume ($)')
plt.grid(True)
plt.show()
```
